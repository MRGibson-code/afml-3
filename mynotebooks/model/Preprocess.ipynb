```python
import pandas as pd

feature = pd.read_csv("../datasets/ADF_adjust_train.csv")
label = pd.read_csv("../datasets/label_train.csv")

feature_label = pd.merge(feature, label, on="trade_date", how="inner")
feature_label.head()
```

```python
cols = feature_label.columns
X = feature_label[cols[2:-3]].values
y = feature_label["label"].values
print(X.shape, y.shape)
```

```python
# 标准化并降维尝试
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

myscaler = StandardScaler()
X_scale = myscaler.fit_transform(X)

mypca = PCA(0.95)
X_pca = mypca.fit_transform(X_scale)
X_pca.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

var_ = mypca.explained_variance_ratio_
plt.plot(range(len(var_)), var_.cumsum(), c='r')
plt.title("PCA:Explained variance ratio")
plt.xlabel("n_components")
plt.show()
```

```python
# meta labeling
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import linear_model
model = linear_model.LogisticRegression(
    multi_class='multinomial', solver='newton-cg')

model.fit(X_pca, y)
y_pred_p = model.predict_proba(X_pca)
y_pred_c = model.predict(X_pca)

precision1_list = []
precision_ne_1_list = []
for c in np.arange(0, 1, 0.01):
    y_pred_c[y_pred_p[:, 0] > y_pred_p[:, 2]] = -1
    y_pred_c[y_pred_p[:, 0] <= y_pred_p[:, 2]] = 1
    y_pred_c[y_pred_p[:, 1] > c] = 0

    TP_1 = y_pred_c[(y_pred_c == 1) & (y == 1)].shape[0]
    P_1 = y_pred_c[y_pred_c == 1].shape[0]

    TP_ne_1 = y_pred_c[(y_pred_c == -1) & (y == -1)].shape[0]
    P_ne_1 = y_pred_c[y_pred_c == -1].shape[0]

    precision1 = TP_1/P_1 if P_1 != 0 else 0
    precision_ne_1 = TP_ne_1/P_ne_1 if P_ne_1 != 0 else 0

    precision1_list.append(precision1)
    precision_ne_1_list.append(precision_ne_1)

plt.figure()
plt.plot(np.arange(0, 1, 0.01), precision1_list, c='r', label="precision of 1")
plt.plot(np.arange(0, 1, 0.01), precision_ne_1_list,
         c='b', label="precision of -1")

plt.legend()
plt.show()
```

```python
p = pd.DataFrame()
p["p1"] = precision1_list
p["p2"] = precision_ne_1_list
p["p1+p2"] = p["p1"]+p["p2"]
p["c"] = np.arange(0, 1, 0.01)
p.head()
```

```python
# 混淆矩阵
c = 0.2
y_pred_c[y_pred_p[:, 0] > y_pred_p[:, 2]] = -1
y_pred_c[y_pred_p[:, 0] <= y_pred_p[:, 2]] = 1
y_pred_c[y_pred_p[:, 1] > c] = 0
print(classification_report(y, y_pred_c))
```

```python
# 最终结果，注意加上trade_date
cols = feature_label.columns
result_df = pd.DataFrame(columns=["PCA_"+str(i)
                                  for i in range(X_pca.shape[1])], data=X_pca)
result_df["y_primary"] = y_pred_c.reshape(-1, 1)
result_df["y_label"] = np.maximum(y*y_pred_c, 0)

result_df["trade_date"] = feature_label["trade_date"]
result_df["FirstTime"] = feature_label["FirstTime"]
result_df.head()
```

```python
tmp = pd.DataFrame()
tmp["one_zero_nums"] = result_df["y_label"].value_counts()
tmp
```

```python
# 结果保存

import pickle

# train set
result_df.to_csv("../datasets/after_meta_train.csv", index=False)

# StandardScaler,PCA,logistic
with open("../datasets/train_pca.pkl", "wb+") as f:
    pickle.dump(mypca, f)
with open("../datasets/train_standardscler.pkl", "wb+") as f:
    pickle.dump(myscaler, f)
with open("../datasets/train_logistic.pkl", "wb+") as f:
    pickle.dump(model, f)  
```
