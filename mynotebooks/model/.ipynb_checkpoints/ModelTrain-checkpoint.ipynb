```python
import pandas as pd

df = pd.read_csv("../datasets/after_meta_train.csv")

df["trade_date"] = pd.to_datetime(df["trade_date"])
df["trade_date"] = pd.to_datetime(df["trade_date"])
df.head()
```

```python
def get_independence(sr1, sr2, date_sr):
    C = pd.DataFrame()
    C["date"] = date_sr
    C["count"] = 0
    C["date"] = pd.to_datetime(C["date"])
    for start_date, end_date in zip(sr1, sr2):
        C.loc[(C["date"] > start_date) & (C["date"] <= end_date), "count"] += 1
    result = pd.DataFrame()
    result["start_date"] = sr1
    for i, (start_date, end_date) in enumerate(zip(sr1, sr2)):
        c = C.loc[(C["date"] > start_date) & (C["date"] <= end_date), "count"]
        if len(c) == 0:
            result.loc[i, "p"] = 1/2
            continue
        result.loc[i, "p"] = ((1/(c+1)).sum())/len(c)
    return result


independence = get_independence(
    df["trade_date"], df["FirstTime"], df["trade_date"])
independence.head()
```

```python
# sample weight
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, roc_auc_score
import warnings

warnings.filterwarnings("ignore")
sample_weight = independence["p"]/sum(independence["p"])

# 5-fold 交叉验证
K = 5
df_len = len(df)//K
start_time_list = [df.loc[df_len*i, "trade_date"]
                   for i in range(K)]+[df.iloc[-1]["trade_date"]]

# 防止信息泄露
date_df = pd.DataFrame()
date_df["trade_date"] = df["trade_date"]
date_df["trade_date_pre_30"] = date_df["trade_date"].shift(30)
date_df["trade_date_30"] = date_df["trade_date"].shift(-30)
date_df.fillna(method="bfill", inplace=True)
date_df.fillna(method="ffill", inplace=True)
date_df.set_index("trade_date", inplace=True)

max_depth_list = [None, 2, 5]
max_features_list = ["auto", 2, 5]
score_list = []
for max_depth in max_depth_list:
    for max_features in max_features_list:
        score = 0
        for i in range(K):
            df_valid = df[(df["trade_date"] > start_time_list[i])
                          & (df["trade_date"] < start_time_list[i+1])]

            X_valid = df_valid.values[:, :-3]
            y_valid = df_valid.values[:, -3].astype("int")

            valid_date_start = df_valid.iloc[0]["trade_date"]
            valid_date_end = df_valid.iloc[-1]["trade_date"]
            df_train = df[(df["trade_date"] < date_df.loc[valid_date_start, "trade_date_pre_30"]) |
                          (df["trade_date"] > date_df.loc[valid_date_end, "trade_date_30"])]

            X_train = df_train.values[:, :-3]
            y_train = df_train.values[:, -3].astype("int")

            sample_weight = pd.merge(df_train, independence,
                                     left_on="trade_date", right_on="start_date", how="inner")["p"]

            # 500棵树
            myforest = RandomForestClassifier(
                max_depth=max_depth, n_estimators=500)
            myforest.fit(X_train, y_train)
            score_ = myforest.score(X_valid, y_valid)
            score += score_
        score_list.append(score/K)

score_list
```

```python
tmp = pd.DataFrame()
i = 0
for max_depth in max_depth_list:
    for max_features in max_features_list:
        tmp.loc[i, "max_depth"] = max_depth
        tmp.loc[i, "max_features"] = max_features
        tmp.loc[i, "score"] = score_list[i]
        i += 1
tmp
```

```python
tmp = pd.DataFrame()
tmp["max_depth"] = max_depth_list
tmp["max_features"] = max_features_list
tmp
```

```python
# 模型训练
max_depth = 5
max_features = "auto"
n_estimators = 1000

X_train = df.values[:, :-3]
y_train = df.values[:, -3].astype("int")
myforest = RandomForestClassifier(
    max_depth=max_depth, n_estimators=n_estimators)
myforest.fit(X_train, y_train)
```

```python
test_df = pd.read_csv("../datasets/after_meta_train.csv")
X_test = test_df.values[:, :-3]
y_test = test_df.values[:, -3].astype("int")

myforest.score(X_test, y_test)
```
